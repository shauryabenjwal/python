# Q4. *Enter a string and substring, then print how many times substring occurs from left or right.*

```python
main_string = "ABCBEFCDCBA"
sub_string = "BCD"

count = 0
start = 0
while True:
    start = main_string.find(sub_string, start)
    if start == -1:
        break
    count += 1
    start = start + 1

print("occurrences: ", count)
```

# Q5. *Count occurences of each alphabet (case-sensitive) in a mixed-case string.*

```python
text = "ABcDE"
frequency = {}

for ch in text:
    if ch.isalpha():
        key = ch.upper()
        frequency[key] = frequency.get(key, 0) + 1

for key in sorted(frequency):
    print(f"{frequency[key]} {key}") 
```

# Q6. *Count number of unique words in a sequence using sets.*

```python
sentence = "python is simple and python is powerful."
words = sentence.lower().split()
unique_words = set(words)

print("Unique words: ", unique_words)
print("count of unique words: ", len(unique_words))
```

# Q7. *Create two sets of fruits and find*
* fruits in both sets.
* fruits only in *s1*.
* Counts of all fruits from both sets.

```python
s1 = {"apple","banana","mango","grapes"}
s2 = {"banana","orange","kiwi","mango"}

print("Fruits in both sets: ", s1 & s2)
print("Fruits only in s1: ", s1 - s2)
print("Count of all unique fruits: ", len(s1 | s2))
```

# Q8. *Take two sets and apply various set operations.*

```python
s1 = {"red","yellow","orange","blue"}
s2 = {"violet","blue","purple"}

print("Union: ", s1 | s2)
print("Intersection: ", s1 & s2)
print("difference (s1 - s2): ", s1 - s2)
print("difference (s2 - s1): ", s2 - s1)
print("symmetric difference: ", s1 ^ s2)
```
