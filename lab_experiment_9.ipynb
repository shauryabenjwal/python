```python
class student:
    def __init__self(self, name,sap_id,phy,chem,maths):
        self.name = name
        self.sap_id = sap_id
        self.phy = phy
        self.chem = chem
        self.maths = maths

    def display(self):
        print("Name:", self.name)
        print("sap_id", self.sap_id)
        print("marks (Phy, Chem, Math):", self.phy, self.chem, self.maths)

students = []

for i in range(3):
    name = input("Enter Name: ")
    sap_id = input("Enter sap_id: ")
    phy = int(input("physics marks: "))
    chem = int(input("chemistry marks: "))
    maths = int(input("mathematics marks: "))

    s = student(name, sap_id, phy, chem, maths)
    students.append(s)

for s in students:
    s.display()
```

```python
class A:
    def funcA(self):
        print("Class A")

class B(A):
    def funcB(self):
        print("Class B")

class C(B):
    def funcC(self):
        print("Class C")

obj = C()
obj.funcA()
obj.funcB()
obj.funcC()
```

* A -> B -> C Chain.

* C can access both A & B.

```python
class A:
    def funcA(self):
        print("Class A")

class B:
    def funcB(self):
        print("Class B")

class C(A, B):
    def funcC(self):
        print("Class C")

obj = C()
obj.funcA()
obj.funcB()
```

* One child inherits from multiple parents.

```python
class parent:
    def show(self):
        print("parent method")

class child(parent):
    def show(self):
        print("child method")

obj = child()
obj.show()
```

* Same method name in parent & child.

* Child method overrides parent method.

* Output -> "Child Output".
