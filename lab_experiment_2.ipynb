# Q8. Find area of a triangle when lengths of all three sides are given *(Heron`s formula)*.

```python
import math 
a, b, c = 3, 4, 5
s = (a+b+c)/2
area = math.sqrt (s * (s-a) * (s-b) *(s-c))
print("area of triangle: ", area)
```
