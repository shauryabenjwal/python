# Q1. *Write a function to find maximum and minimum from a sequence without using built-in *max()* and *min()* *

```python
def find_max_min(sequence):
    maximum = sequence[0]
    minimum = sequence[0]

    for value in sequence[1:]:
        if value > maximum:
            maximum = value
        if value < minimum:
            minimum = value

    return maximum, minimum

nums = [15, 3, 99, 42, 7]
mx, mn = find_max_min(nums)
print("Maximum:", mx)
print("Minimum:", mn)
```

# Q2. *Write a function that takes a positive integer and returns sum of cubes of all positive integers smaller than it.*

```python
def sum_of_cubes_below(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

n = 6
print(f"Sum of cubes below {n}:", sum_of_cubes_below(n))
```

# Q3. *Write a recursive function to print numbers from 1 to n (no loops).*

```python
def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n - 1)
    print(n, end=" ")

print_1_to_n(10)
print()
```

# Q4. *Write a recursive function to print Fibonacci series up to n terms.*

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

terms = 8
print("Fibonacci series:")
for i in range(terms):
    print(fib(i), end=" ")
print()
```

# Q5. *Write a lambda function to find volume of a cone.*

```python
import math

cone_volume = lambda r, h: (1 / 3) * math.pi * r ** 2 * h

print("Volume of cone:", cone_volume(3, 5))
```

# Q6. *Write a lambda function to return tuple of maximum and minimum from a list.*

```python
data = [10, 6, 8, 90, 12, 56]
max_min_tuple = lambda values: (max(values), min(values))

print("Result:", max_min_tuple(data))
```
