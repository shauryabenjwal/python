# Q5. *Check whether a quadratic equation has real or imaginary roots and display their roots.*

```python
import cmath 

a, b, c =1, 2, 5

discriminant = b**2 - 4*a*c
root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
root2 = (-b - cmath.sqrt(discriminant)) / (2*a)

if discriminant > 0:
    print("roots are equal")
else:
    print("roots are imaginary")

print("root1:", root1)
print("root2:", root2)
```

# Q6. *Find whether a given year is a leap years or not.*

```python
year = 2024

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year,"is not a leap year")    
```
