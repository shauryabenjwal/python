# Q1. Add few names, one name in each row, in "name.txt.file".

* Count no of names.
* Count all names starting with vowel.
* Find longest name.

```python
with open("name.txt", "w") as file:
    file.write("Aman\nOmen\nskye\nJett\n")
```

explanation 

* "w" -> write mode (creates/overwrites file).
* \n -> moves to next line *(one name per row)*.

```python
with open("name.txt", "r") as file:
    names = file.read().splitlines()
    print("total names:",  len(names))
```

explanation

* "r" -> read mode.
* readlines() -> returns list of lines.
* len(names) -> counts total names.

```python
count = 0
vowels = "aeiouAEIOU"

with open("name.txt", "r") as file:
    for name in file:
        if name.strip()[0] in vowels:
            count += 1

print("Names starting with a vowel:", count)
```

explanation

* *strip()* -> removes spaces/newline.
* *[0]* -> first character.
* Check if first letter is vowel.

```python
longest = ""
with open("name.txt", "r") as file:
    for name in file:
        if len(name.strip()) > len(longest):
            longest = name

print("longest name:", longest)
```

* Compare length of each name.
* Store longest found.

```python
n = int(input("enter a number of test cases: "))

for i in range(n):
    try:
        a, b = input().split()
        result = int(a)  // int(b)
        print(result)

    except ZeroDivisionError as e:
        print("Error: Division by zero is not allowed.", e)
    
    except ValueError as e:
        print("Error : Invalid input. Please enter two integers. ", e)
```

* try -> Code that may cause error.
* except -> Handles errors.

```python
class FileEmptyError(Exception):
    pass

class InvalidDataError(Exception):
    pass

try:
    with open("numbers.txt", "r") as file:
        data = file.read().strip()

        if data == "":
            raise FileEmptyError("The file is empty.")
        
        for line in data.split("\n"):
            if not line.isdigit():
                raise InvalidDataError("Invalid data found")

except FileNotFoundError:
    print("File not found")

except FileEmptyError as e:
    print(e)

except InvalidDataError as e:
    print(e)
```
